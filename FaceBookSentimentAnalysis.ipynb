```python
import pandas as pd
from textblob import TextBlob

# Step 1: Create dataset with 3 columns
data = {
    "user": [
        "Arun Kumar","Priya Sharma","Rahul Verma","Sneha Iyer","Vikram Singh",
        "Anjali Mehta","Karthik Raj","Neha Gupta","Rohan Das","Divya Nair"
    ],
    "comment": [
        "This product is amazing!",
        "It is okay, nothing special",
        "Worst experience ever",
        "Average performance",
        "I love it",
        "Not bad, not great",
        "Could be better",
        "Absolutely fantastic!",
        "It works fine",
        "Service was neither good nor bad"
    ],
    "date": [
        "2026-01-01","2026-01-02","2026-01-03","2026-01-04","2026-01-05",
        "2026-01-06","2026-01-07","2026-01-08","2026-01-09","2026-01-10"
    ]
}

df = pd.DataFrame(data)
df
```

```python
# Step 2: Sentiment function with neutral range
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity

    if polarity > 0.1:
        return "Positive"
    elif polarity < -0.1:
        return "Negative"
    else:
        return "Neutral"
```

```python
# Apply sentiment
df["Sentiment"] = df["comment"].apply(get_sentiment)

# Step 3: Filter only Neutral feedback
neutral_df = df[df["Sentiment"] == "Neutral"]

print("\nNeutral Feedback:")
print(neutral_df)
```
